## Tracebase Analysis
This jupyter presents the time series plots of all lamps provides by the Tracebase dataset. The dataset contains one day of measurements with a sampling rate of 1 measurement/second for multiple appliances. A histogram plot is also presented.

```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
all_files = os.listdir('./complete/Lamp')
```

```python
csv_path = './complete/Lamp/{}'.format(all_files[0])
```

```python
df = pd.read_csv(csv_path, sep=';', header=None, names=['Date', '1sec', '8sec'])
```

```python
for file_name in all_files:
    csv_path = './complete/Lamp/{}'.format(file_name)
    df = pd.read_csv(csv_path, sep=';', header=None, names=['Date', '1sec', '8sec'])
    plt.figure(figsize=(16,3))
    plt.plot(df['1sec'])
    plt.title('File name: {}'.format(file_name))
```

```python
n_graphs = len(all_files)
plt.figure(figsize=(16,3*n_graphs))
for i, file_name in enumerate(all_files):
    csv_path = './complete/Lamp/{}'.format(file_name)
    df = pd.read_csv(csv_path, sep=';', header=None, names=['Date', '1sec', '8sec'])
    plt.subplot(n_graphs/2,2,i+1)
    df[df['1sec']>5]['1sec'].hist(bins=20)
    plt.title('File name: {}'.format(file_name))
```
